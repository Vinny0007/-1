```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify-2023.csv",encoding="latin-1") #довелось зазначити кодування тому що pandas не міг прочитати файл
```

```python
#Завдання 1
#Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.
df = df[df["released_year"] == 2023]
df = df.sort_values(by="streams", ascending=False)
top_10_songs = df.head(10)
top_10_songs = top_10_songs[::-1]
# Побудова  діаграми
plt.figure(figsize=(15, 9))
plt.barh(top_10_songs["track_name"], top_10_songs["streams"], color="skyblue")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Назва пісні")
plt.title("10 найпопулярніших пісень на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```

```python
#Завдання 2
songs_2023 = df[df["released_year"] == 2023]

grouped_data = songs_2023.groupby("released_month").size()


months = [
    "Січень", "Лютий", "Березень", "Квітень", "Травень", "Червень",
    "Липень", "Серпень", "Вересень", "Жовтень", "Листопад", "Грудень"
]


plt.figure(figsize=(8, 8))
plt.pie(grouped_data, autopct='%1.1f%%', startangle=140, pctdistance=0.85)


plt.legend(months, title="Місяці", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title("Розподіл пісень за місяцем випуску у 2023 році")

plt.show()
```

```python
#Завдання 3

selected_columns = ["track_name", "released_month", "streams"]
songs_2023_selected = songs_2023[selected_columns]


grouped_data = songs_2023_selected.groupby("track_name")


plt.figure(figsize=(12, 6))

for track_name, track_data in grouped_data:
    plt.plot(
        track_data['released_month'],
        track_data['streams'],
        marker='o',
        linestyle='-',
        label=track_name
    )

plt.xlabel("Місяць випуску")
plt.ylabel("Кількість прослуховувань")
plt.title("Динаміка популярності треків у 2023 році")
plt.xticks(songs_2023_selected["released_month"].unique())
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1), fontsize='small')
plt.grid(True)
plt.show()
```

```python
#Завдання 4
df = df.sort_values(by='streams', ascending=False)


top_10_artists = df.head(10)


plt.figure(figsize=(12, 6))
plt.barh(top_10_artists['artist(s)_name'], top_10_artists['streams'], color="blue")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Артисти")
plt.title("10 найпопулярніших артистів на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Завдання 5
tracks = df.sample(n=3)
plt.figure(figsize=(10, 4))
plt.barh(tracks["track_name"], tracks["in_spotify_playlists"], color="grey")
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Назва треку")
plt.title("Порівняння популярності трьох треків на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('shopping_trends.csv')


print("Розмір таблиці:", df.shape)


print("Наявність пропусків:")
print(df.isnull().sum())


print("Типи стовпців:")
print(df.dtypes)


df.columns = df.columns.str.lower().str.replace(' ', '_')


print("Оновлені назви стовпців:")
print(df.columns)
```

```python
#Завдання 2.1

sns.histplot(df['age'], bins=20, kde=True, color='green')  

plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Кількість клієнтів')
plt.show()
```

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn.

Налаштуйте відображення графіка, включаючи назву графіку, підписи осей і інші необхідні параметри для кращого розуміння візуалізації.

Додайте кольорову лег
енду.

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
surface = ax.plot_surface(x, y, z, cmap="plasma")
ax.set_title("3D Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10) 
plt.show()
```
