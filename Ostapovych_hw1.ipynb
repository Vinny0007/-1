```python
#Завдання 5 створіть два одновимірні масиви розміром 6, заповніть його випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення
vector_1 = np.array([np.random.randint(1, 21) for n in range(6)])
vector_2 = np.array([np.random.randint(1, 21) for n in range(6)])
print(vector_1 - vector_2)
print(vector_1 + vector_2)
print(vector_1 * vector_2)
```

```python
#Завдання 6 створіть дві матриці розміром 3х2 та 2х4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою
matrix_1 = np.random.randint(1, 6, size=(3, 2))
matrix_2 = np.random.randint(1, 6, size=(2, 4))
product = np.matmul(matrix_1, matrix_2)
print(product)
```

```python
#Завдання 7 створіть матрицю розміром 3х3 заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її 
matrix = np.array(np.random.randint(2, size=(3, 3)))
transp_matrix = np.transpose(matrix)
print(matrix)
print(transp_matrix)
```

```python
#Завдання 9 створіть дві матиці розміром 3х3,заповніть їх випадковими дійсними числами в діапазоні від -5 до 5 та знайдіть їх добуток
matrix_1 = np.array(np.random.randint(-5, 6,  size=(3, 3)))  
matrix_2 = np.array(np.random.randint(-5, 6,  size=(3, 3)))
product = np.matmul(matrix_1, matrix_2) 
print(product)
```

```python
#Завдання 10 створіть матрицю розміром 3х3 заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик що містить суму елементів кожного рядка матриці
matrix = np.array(np.random.uniform(-1, 1, 9)).reshape(3, 3)
vector = np.array([matrix[0].sum(), matrix[1].sum(), matrix[2].sum()])
r_vector = vector.reshape(3, 1)
print(r_vector)
```

```python
url = ("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96")
data = pd.read_html(url, header=0)[20]
print(data)
```

```python
#Завдання 7 Видалити дані для всієї країни з таблиці в останньому рядку таблиці 
data_d = data.drop([27])
data_d
```

```python
#Завдання 9 Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні 
data_h = data_d[data_d["2014"] > data_d["2014"].mean()]
print(data_d["2014"].mean())
data_h[["регіон", "2014"]]
```
